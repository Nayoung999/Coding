# Unit 18. break, continue로 반복문 제어하기 

- break는 for와 while 문법에서 제어흐름을 벗어나기 위해 사용한다. 
- 즉, 루프를 완전히 중단한다. 
- continue는 제어흐름(반복)을 유지한 상태에서 코드의 실행만 건너뛰는 역할을 한다. 

-----

- break : 제어흐름 중단
- continue : 제어흐름 유지, 코드 실행만 건너 뜀



````python
# 18.1 break으로 반복문 끝내기
# 18.1.1 while에서 break으로 반복문 끝내기 
# while 무한 루프에서 숫자를 증가시키다가 변수 i 가 100일 때, 반복문을 끝내자. 

i = 0
while True: # 무한루프
    print(i)
    i += 1
    if i == 100:
        break  # 반복문을 끝냄, while의 제어 흐름을 벗어난다. 
````

- memo : break에는 :(콜론)을 붙이지 않는다. 

````python
# 18.1.2 for에서 break로 반복문 끝내기

for i in range(10000): #0~9999
    print(i)
    if i == 100: # i가 100일 때
        break  # 반복문을 끝낸다. for의 제어흐름을 벗어난다. 
````

````python
# 18.2 continue로 코드 실행 건너뛰기

# 18.2.1 for에서 continue로 코드 실행 건너뛰기 

for i in range(100): #0부터 99까지 증가하면서 100번 반복
    if i % 2 ==0: # i를 2로 나누었 을 때 나머지가 0이면 짝수다.
        continue  # 아래 코드를 실행하지 않고 건너 뛴다. 
    print(i)
````

````python
# memo : continue에도 :(콜론)을 붙이지 않는다.
````

````python
# 18.2.2. while 반복문에서 continue로 코드 실행 건너뛰기

i = 0
while i < 100: # i가 100보다 작을 때로 반복횟수를 지정함
              # 0부터 99까지 증가하면서 100번 반복
    i =+ 1 #i를 1씩 증가시킴
    if i % 2 == 0:  # i를 2로 나누었을 때 나머지가 0이면 짝수
        continue    # 아래 코드를 실행하지 않고 지나감 
    print(i)
````

````python

# 18.3 입력한 횟수대로 반복하기 

count = int(input('반복할 횟수를 입력하세요: '))

i = 0
while True: # 무한 루프
    print(i)
    i += 1
    if i == count : # i가 입력받은 값과 같을 때
        break       # 반복문을 끝낸다. 



# input으로 입력값을 받아서 count 변수에 저장했다. 
# 이때, 반드시 int를 사용해서 input에서 나온 문자열을 정수로 변환해준다. 
# i에는 0을 할당하고, while에 True를 지정하여 무한 루프로 만든다. 
# 반복문 안에서는 i의 값을 출력하고, 변화식에서는 i를 1씩 증가시킨다. 
# i가 count의 값과 같으면 break를 실행한다.
````

````python
# 18.3.1 입력한 숫자까지 홀수 출력하기 

count = int(input('반복할 횟수를 입력하세요: '))

for i in range(count + 1):   # 0부터 증가하면서 count까지 반복한다. 
    if i % 2 == 0:  # i를 2로 나누었을 때의 나머지가 0이면 i는 짝수
        continue
    print(i)
````

- for의 range에 count+1을 지정하는 이유
range(count)는 0부터 시작하기 때문에 count의 값은 반복에 포함되지 않는다. 따라서 count에 1을 더해서 count의 값도 반복에 포함되도록 만들어준다.

````python
# 18.5 연습문제 : 3으로 끝나는 숫자만 출력하기 
# 0과 73 사이의 숫자 중 3으로 끝나는 숫자만 출력되게 만드세요. 
# 다시 하기 

i = 0
while True:
  if i % 10 != 3:
      i += 1
      continue

  if i > 73: # 73을 포함해야하니까 73보다 숫자가 큰 경우만 cut
    break

  print(i, end=' ')
  i += 1
````

# Unit 19. 계단식으로 별 출력하기 
- for 반복문과 if 조건문을 사용하여 계단식을 별(*)을 출력해보자. 

````python
# 19.1 중첩 루프 사용하기 
# 콘솔은 2차원 평면이다. 별을 일정한 모양으로 출력하려면 반복문을 두 개 사용하는 것이 편리하다. 
# 반복문 안에 반복문이 들어가는 형태를 중첩 루프라고 한다. 
# 반복문의 루프 인덱스 변수는 index의 첫글자를 따서 i를 사용한다. 
# i를 사용하는 바깥쪽 루프는 세로 방향을 처리하고, j를 사용하는 안쪽 루프는 가로 방향을 처리한다. 

for i in range(5): # 5회 반복, 바깥쪽 루프는 세로 방향
    for j in range(5):  # 5회 반복, 안쪽 루프는 가로 방향
      print('j:', j, sep='', end= ' ') # j값 출력, end에 ' '를 지정하여 줄바꿈 대신 한 칸 띄운다. 
    print('i:', i, '\\n', sep= '') # i값 출력, 개행 문자 모양도 출력
    # 가로 방향으로 숫자를 모두 출력한 뒤 다음 줄로 넘어간다. 
    # print는 기본적으로 출력 후 다음 줄로 넘어감 

# 중첩 루프는 2차원 평면을 다룰 수 있으므로 이미지 처리, 영상 처리, 좌표계 처리 등에 주로 쓰인다. 
````

````python
# 19.2 사각형으로 별 출력하기 

for i in range(5):  #바깥 루프는 세로 
    for j in range(5): # 안쪽 루프는 가로
        print('*', end='')  # * print, end에 ''를 지정해서 줄바꿈을 하지 않는다. 
    print() #가로 방향으로 별을 다 그린 뒤 다음 줄로 넘어간다. 
            # print는 출력 후 기본적으로 다음 줄로 넘어간다. 
````

````python
# 19.2.1 사각형 모양 바꾸기 
# for 반복문의 조건식을 수정하여 사각형의 모양을 바꿔준다. 

for i in range(3):
    for j in range(7):
        print('*', end='') # * print, end = ''을 통해서 줄바꿈을 하지 않는다. 
        # print는 출력 후 기본적으로 다음 줄로 넘어가는데 그것을 막음.
    print() 
````

````python
# 19.3 계단식으로 별 출력하기 

for i in range(5):
    for j in range(5):
        if j <= i:  #세로 방향 변수 i만큼
            print('*', end='')  # 별을 출력한다. end에 ''을 지정하여 줇바꿈 해제?
    print()
````

````python
# 19.3.1 대각선으로 별 출력하기 
for i in range(5):
    for j in range(5):
        if j ==i:
            print('*', end='')
    print()
````

````python
# 대각선을 출력하려고 했는데 안된다? why?
# IDLE이나 콘솔에서는 문자 출력을 하지 않으면, 다음 문자는 현재 위치에 출력되므로 모두 왼쪽으로 몰린다. 
# 따라서 별을 출력하지 않는 부분에서는 print로 공백을 출력해주어야 별이 원하는 위치에 오게 된다. 

for i in range(5):
    for j in range(5):
        if j == i:    # 세로 방향의 변수와 같을 때 
            print('*', end='')  # 별을 출력한다. end에 ''을 지정하여 줄바꿈을 하지 않는다. 
        else:   # 세로 방향의 변수와 다를 때
            print(' ', end='')  # 공백 출력, end 에 ''을 지정하여 줄바꿈을 하지 않는다. 

    print() # 가로 방향으로 별을 다 그린 후에 다음 줄로 넘어 간다. 
````

````python
# 19.5 연습문제 : 역삼각형 모양으로 별 출력하기 
# memo : 다시 풀기. 스스로 하기 힘들다. 

for i in range(5):
    for j in range(5):
        if j < i :
            print(' ', end='')
        else:
            print('*', end='')
    print()

# 공백 부분이 세로 방향인 줄의 위치에 비례하므로 조건식도 세로 방향인 변수 j에 맞추어 작성한다. 
# 단, 여기서는 i보다 j가 1이 작은 상황이다. 
# 따라서 if 조건문은 j < i :와 같이 작성해서 가로 방향으로 공백을 출력할 때는 세로 방향의 변수 i보다 작을 때까지 출력하고, 반대되는 조건에서는 별을 출력하면 된다. 
````

# Unit 20. FizzBuzz 문제

- FizzBuzz는 매우 간단한 프로그래밍 문제이며 규칙은 다음과 같습니다.

- 1에서 100까지 출력
- 3의 배수는 Fizz 출력
- 5의 배수는 Buzz 출력
- 3과 5의 공배수는 FizzBuzz 출력
- 즉, 1부터 100까지 숫자를 출력하면서 3의 배수는 숫자 대신 'Fizz', 5의 배수는 숫자 대신 'Buzz', 3과 5의 공배수는 숫자 대신 'FizzBuzz'를 출력하면 됩니다.

````python
# 20.1. 1부터 100까지 숫자 출력하기 
# 반복문, 조건문, 나머지 연산자, 비교 연산자 모두를 동원해야 풀 수 있다.  
for i in range(1, 101):
    print(i)
````

````python
# 20.2 3의 배수일 때와 5의 배수일 때 처리하기 
for i in range(1, 101):
    if i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
````
